# ***Scraping reddit posts for player performance analysis***

```python
import csv
import praw
from datetime import datetime

# Reddit API credentials
client_id = "eav4DqQgy4X2pnahats03Q"
client_secret = "LPylQNpAbExEsk8kMT0VQbUvEFpfRQ"
user_agent = "Cold-Forever-1798"

# Connecting to Reddit
reddit = praw.Reddit(client_id=client_id,
                     client_secret=client_secret,
                     user_agent=user_agent)

# List of players with their best and worst seasons
players = {
    "Marco Asensio": {
        "best": (2022, 2023),
        "worst": (2019, 2020)
    },
    "Vinicius jr": {
        "best": (2021, 2022),
        "worst": (2019, 2020)
    },
    "Rashford": {
        "best": (2019, 2020),
        "worst": (2021, 2022)
    },
    "Neymar": {
        "best": (2014, 2015),
        "worst": (2021, 2022)
    },
    "Eden Hazard": {
        "best": (2016, 2017),
        "worst": (2020, 2021)
    },
    "Ousmane Dembele": {
        "best": (2016, 2016),
        "worst": (2018, 2019)
    },
    "Raphinha": {
        "best": (2024, 2025),
        "worst": (2022, 2023)
    },
    "Gareth Bale": {
        "best": (2012, 2013),
        "worst": (2017, 2018)
    },
    "Leroy Sané": {
        "best": (2017, 2018),
        "worst": (2019, 2020)
    }
}

limit = 250  # Maximum number of posts to fetch per search

# Opening a csv file to store the data scraped
with open("reddit_posts_players.csv", "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["player", "season_type", "season", "title", "posted_on", "subreddit", "score", "url"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    # Looping through each player and their best and worst season
    for player, seasons in players.items():
        for season_type, (start_year, end_year) in seasons.items():
            # Changing season years to timestamps
            start_date = datetime(start_year, 1, 1)
            end_date = datetime(end_year, 12, 31, 23, 59, 59)
            start_timestamp = start_date.timestamp()
            end_timestamp = end_date.timestamp()

            # Searching for reddit posts mentioning the players names
            query = player
            posts = reddit.subreddit("all").search(query, limit=limit)

            # Saving posts in seasons time range
            for post in posts:
                if start_timestamp <= post.created_utc <= end_timestamp:
                    post_time = datetime.utcfromtimestamp(post.created_utc).strftime('%Y-%m-%d %H:%M:%S')
                    writer.writerow({
                        "player": player,
                        "season_type": season_type,
                        "season": f"{start_year}-{end_year}",
                        "title": post.title,
                        "posted_on": post_time,
                        "subreddit": post.subreddit.display_name,
                        "score": post.score,
                        "url": post.url
                    })

print("CSV file 'reddit_posts_players.csv' has been created.")
```

# ***Converting audio to text***

```python
import speech_recognition as sr

def transcribe_audio(audio_file_path):
    # Creating a recognizer
    recognizer = sr.Recognizer()
    # Loading the audio file
    with sr.AudioFile(audio_file_path) as source:
        audio = recognizer.record(source)
    try:

        text = recognizer.recognize_google(audio)
        return text
    except sr.RequestError:
        return "API unavailable."
    except sr.UnknownValueError:
        return "Unable to recognize speech."
```

```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin

# === CONFIG ===
url = "https://www.yumpu.com/en/document/view/69250688/read-pdf-raphinha-brazilamp039s-bright-star-lighting-up-the-soccer-world-for-kids-pdf-ebook"
download_folder = "pdf_downloads"

# === Create folder to save PDFs ===
os.makedirs(download_folder, exist_ok=True)

# === Fetch and parse the webpage ===
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# === Find all PDF links ===
pdf_links = []
for link in soup.find_all("a", href=True):
    href = link["href"]
    if href.lower().endswith(".pdf"):
        full_url = urljoin(url, href)  # Handle relative links
        pdf_links.append(full_url)

print(f" Found {len(pdf_links)} PDF files.")

# === Download the PDFs ===
for i, pdf_url in enumerate(pdf_links, start=1):
    try:
        file_name = os.path.join(download_folder, os.path.basename(pdf_url))
        print(f"⬇️  Downloading {pdf_url}")
        pdf_response = requests.get(pdf_url)
        with open(file_name, 'wb') as f:
            f.write(pdf_response.content)
        print(f" Saved: {file_name}")
    except Exception as e:
        print(f" Failed to download {pdf_url}: {e}")
```

```python
import os
from pdfminer.high_level import extract_text
from pdf2image import convert_from_path
import pytesseract

# List of PDF file paths
pdf_paths = [
    '/content/Oussmane dembele (1).pdf',
    '/content/pdf.raphinha.pdf',
    '/content/Marco asensio.pdf',
    '/content/Leroy sane.pdf',
    '/content/Young_Kings_Marcus_Rashford_and_Theopolitical_Char.pdf',
    '/content/603010077-Neymar-Jr.pdf'
]

# Process each PDF file individually
for pdf_path in pdf_paths:
    if not os.path.exists(pdf_path):
        print(f"File not found: {pdf_path}")
        continue

    pdf_file = os.path.basename(pdf_path)
    print(f"\nProcessing file: {pdf_file}")
    extracted_text_combined = ""

    try:
        extracted_text = extract_text(pdf_path)
    except Exception as e:
        print(f"Error extracting text from {pdf_file} using pdfminer.six: {e}")
        extracted_text = ""

    # If little or no text is extracted, use OCR page by page
    if not extracted_text.strip() or len(extracted_text.strip()) < 10:
        print(f"Insufficient text extracted from {pdf_file}. Switching to OCR...")
        try:
            pages = convert_from_path(pdf_path, dpi=300)
            for i, page in enumerate(pages):
                page_text = pytesseract.image_to_string(page).strip()
                extracted_text_combined += f"Page {i+1}:\n{page_text}\n\n"
                print(f"Processed page {i+1} of {pdf_file} (OCR).")
        except Exception as e:
            print(f"Error processing {pdf_file} with OCR: {e}")
    else:
        print(f"Text extraction successful for {pdf_file}.")
        # Use the text extracted as a whole
        extracted_text_combined = extracted_text.strip()
        print(f"Processed {pdf_file} as one block of text.")

    # Save the extracted text to a separate file
    output_filename = f"extracted_{os.path.splitext(pdf_file)[0]}.txt"
    try:
        with open(output_filename, "w", encoding="utf-8") as file:
            file.write(extracted_text_combined)
        print(f"Extracted text saved to '{output_filename}'.")
    except Exception as e:
        print(f"Error saving the extracted text for {pdf_file}: {e}")

```

# ***Extracting text from PDFs that wasn't extracted***

```python
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
import os

# Setting the Tesseract executable path
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# PDF file
pdf_files = ["/content/Marco asensio.pdf"]

# Extracting text from PDF
def ocr_pdf(pdf_path):
    try:
        # Converting PDF to images
        pages = convert_from_path(pdf_path, 300)
        text_from_all_pages = []

        # Extracting text from each page
        for page_number, page_image in enumerate(pages):
            text = pytesseract.image_to_string(page_image)
            text_from_all_pages.append(text)
            print(f"Processed page {page_number + 1} of {pdf_path}")

        # Combining text from all pages
        return "\n\n".join(text_from_all_pages)
    except Exception as e:
        print(f"Error processing {pdf_path}: {str(e)}")
        return ""

# Processing each PDF file
for pdf_file in pdf_files:
    if not os.path.exists(pdf_file):
        print(f"File not found: {pdf_file}")
        continue

    print(f"\nProcessing PDF file: {pdf_file}")

    # Extracting text using OCR
    extracted_text = ocr_pdf(pdf_file)

    # Saving the extracted text to a file
    text_file_path = pdf_file.replace('.pdf', '_extracted.txt')
    with open(text_file_path, 'w', encoding='utf-8') as f:
        f.write(extracted_text)

    print(f"Text saved to {text_file_path}")
```

# ***Audio Processing and Speech Recognition***

```python
from pydub import AudioSegment
import speech_recognition as sr
import os

# Converting MP3 to WAV
def convert_mp3_to_wav(mp3_file_path, wav_file_path='output_audio.wav'):
    audio = AudioSegment.from_mp3(mp3_file_path)
    audio.export(wav_file_path, format='wav')
    return wav_file_path

# Convert WAV audio to text in 30-second chunks
def speech_to_text(wav_file_path):
    recognizer = sr.Recognizer()
    audio = AudioSegment.from_wav(wav_file_path)
    duration = len(audio)
    chunk_length = 30000
    text = ""

    for i in range(0, duration, chunk_length):
        chunk = audio[i:i+chunk_length]
        chunk.export("temp_chunk.wav", format="wav")
        with sr.AudioFile("temp_chunk.wav") as source:
            audio_data = recognizer.record(source)
        try:
            text += recognizer.recognize_google(audio_data) + " "
        except sr.UnknownValueError:
            text += "(inaudible) "
        except sr.RequestError as e:
            text += "(request error) "

    os.remove("temp_chunk.wav")
    return text

# Speech recognition
mp3_file_path = '/content/podcast.mp3'
wav_file_path = convert_mp3_to_wav(mp3_file_path)
recognized_text = speech_to_text(wav_file_path)

# Print the extracted text
print(recognized_text)


```

# ***Text cleaning and tokenization***

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Downloading required datasets
nltk.download('punkt')
nltk.download('stopwords')

# Function to clean and tokenize text
def clean_and_tokenize(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().lower()

    # Tokenizing text
    tokens = word_tokenize(text)

    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words and word.isalpha()]

    return filtered_tokens

# Cleaning and tokenizing the text
cleaned_tokens = clean_and_tokenize(text_file_path)
```

# ***Automated Sentiment Labeling for Player Reviews***

```python
import pandas as pd
import ast
import os

def label_file(input_file, output_file, threshold=-0.5):
    """
    Loads a CSV file, extracts the 'compound' sentiment score from the 'Sentiment' column,
    and creates a new column 'harsh_label': 1 if compound <= threshold, else 0.
    The labeled DataFrame is saved to output_file.
    """
    # Load the data from a CSV file
    df = pd.read_csv(input_file)
    def get_compound(sent_str):
        try:
            sent_dict = ast.literal_eval(sent_str)
            return sent_dict.get("compound", 0)
        except Exception:
            return 0

    df["compound"] = df["Sentiment"].apply(get_compound)
    df["harsh_label"] = df["compound"].apply(lambda x: 1 if x <= threshold else 0)

    df.to_csv(output_file, index=False)
    print(f"Labeled file saved to: {output_file}")

# List of files to proce
file_list = [
    "/content/processed_Lemmatization_extracted_603010077-Neymar-Jr (1).csv",
     "/content/processed_Lemmatization_Leroy sane_extracted.csv",
    "/content/processed_Lemmatization_Marco asensio_extracted.csv",
    "/content/processed_Lemmatization_Oussmane dembele_extracted.csv",
    "/content/processed_Lemmatization_extracted_Young_Kings_Marcus_Rashford_and_Theopolitical_Char (1).csv",
    "/content/processed_Lemmatization_extracted_pdf.raphinha (1).csv",

]

# Process each file and save labeled version with a suffix "_labeled"
for file_path in file_list:
    base, ext = os.path.splitext(file_path)
    output_file = base + "_labeled" + ext
    label_file(file_path, output_file)
```
